```python
stock = pd.read_csv("../data/US/1m/sp500/ALL.csv")
print(stock.shape)
stock
```

```python
# Convert date column to datetime
stock["date"] = pd.to_datetime(stock["date"])

# Get daily first and last prices
stock["date_only"] = stock["date"].dt.date
daily_data = (
    stock.groupby("date_only")
    .agg({"open": "first", "close": "last", "ticker": "first"})
    .reset_index()
)
daily_data
```

```python
daily_variations = pd.read_csv("../data/US/processed/daily_variations.csv")
print(daily_variations.shape)
daily_variations
```

```python
print(daily_variations.loc[daily_variations["date_only"] == "2024-01-04"].shape)
daily_variations.loc[daily_variations["date_only"] == "2024-01-04"]
```

```python
# Assuming your dataframe is called daily_variations
date = "2024-11-06"
date_data = daily_variations[daily_variations["date_only"] == date]

# Get top positive variations
top_positive = date_data.nlargest(10, "variation")[["ticker", "variation", "open", "close"]]

# Get top negative variations
top_negative = date_data.nsmallest(10, "variation")[["ticker", "variation", "open", "close"]]

print("Top 10 Positive Variations:")
print(top_positive)
print("\nTop 10 Negative Variations:")
print(top_negative)
```

```python
date = "2024-11-06"
ticker = "TSLA"

stock_data = daily_variations[
    (daily_variations["date_only"] == date) & 
    (daily_variations["ticker"] == ticker)
]

if not stock_data.empty:
    variation = stock_data.iloc[0]["variation"]
    open_price = stock_data.iloc[0]["open"]
    close_price = stock_data.iloc[0]["close"]
    print(f"{ticker} on {date}:")
    print(f"Opening: ${open_price:.2f}")
    print(f"Closing: ${close_price:.2f}")
    print(f"Variation: {variation:.2f}%")
else:
    print(f"No data found for {ticker} on {date}")
```
